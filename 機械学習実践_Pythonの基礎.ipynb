```python
def count_9(data):
  count = 0
  for num in data:
    if num % 9 == 0:
      count += 1
  return count


data = [15, 4, 27, 29, 9]
print(count_9(data))
```

```python
def count_35(data):
  count = 0
  for num in data:
    if num % 3 == 0 and num % 5 != 0:
      count += 1
  return count

data = [86, 533, 858, 343, 425, 709, 612, 158, 433, 562, 674, 426, 289, 580, 566]
print(count_35(data))
```
